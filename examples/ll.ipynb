```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
%matplotlib inline
```

```python
qr = QuantumRegister(1)  # register will store one qubit
cr = ClassicalRegister(1)  # register to store one classical bit
circuit = QuantumCircuit(qr, cr)  # create quantum circuit
circuit.h(qr)  # apply H gate to qubit - puts qubit in superposition
circuit.measure(qr, cr)  # measure qubit and store in classical register
```

```python
simulator = Aer.get_backend('qasm_simulator')  # create simulator
result = execute(circuit, backend = simulator, shots = 102400).result()  # execute simulator 1024 times
plot_histogram(result.get_counts())  #plot counts of all executions
```

```python
print(result.get_counts())  # prints out results
```

```python
# |0> result

qr = QuantumRegister(1)  # register will store one qubit
cr = ClassicalRegister(1)  # register to store one classical bit
circuit = QuantumCircuit(qr, cr)  # create quantum circuit

simulator = Aer.get_backend('statevector_simulator')  # create simulator returning state vector of qubit
result = execute(circuit, backend = simulator, shots = 1).result()  # only single shot needed
statevector = result.get_statevector()
print(statevector)  # results in Statevector([1.+0.j, 0.+0.j],dims=(2,))
# j = imaginary unit (i)  so [1, 0] = |0>
array_to_latex(statevector)  # prints state vector => [1 0]
plot_bloch_multivector(statevector)  # print bloch sphere
```

```python
# |+> result without measurement

qr = QuantumRegister(1)  # register will store one qubit
cr = ClassicalRegister(1)  # register to store one classical bit
circuit = QuantumCircuit(qr, cr)  # create quantum circuit
circuit.h(qr)  # apply H gate to qubit - puts qubit in superposition


simulator = Aer.get_backend('statevector_simulator')  # create simulator returning state vector of qubit
result = execute(circuit, backend = simulator, shots = 1).result()  # only single shot needed
statevector = result.get_statevector()
print(statevector)  # results in Statevector([1.+0.j, 0.+0.j],dims=(2,))
# j = imaginary unit (i)  so [1, 0] = |0>
array_to_latex(statevector)  # prints state vector => [1/sqrt(2) 1/sqrt(2)]
plot_bloch_multivector(statevector)  # print bloch sphere
```

```python
# 2 qubit uniform superposition

qr = QuantumRegister(2)  # register will store two qubits
cr = ClassicalRegister(2)  # register to store two classical bits
circuit = QuantumCircuit(qr, cr)  # create quantum circuit
circuit.h(qr)  # apply H gate to qubit - puts qubit in superposition
circuit.measure(qr, cr)


simulator = Aer.get_backend('statevector_simulator')  # create simulator returning state vector of qubit
result = execute(circuit, backend = simulator, shots = 1).result()  # only single shot needed
statevector = result.get_statevector()
print(statevector)
# j = imaginary unit (i)  so [1, 0] = |0>
array_to_latex(statevector)  # prints state vector => |0> or |1>
plot_bloch_multivector(statevector)  # print bloch sphere
```

```python
# 2 qubits - qubit 0 in superposition - no measurement - statevector simulator
# Qubit 0 always |+>, Qubit 1 always |0>
# notice no QuantumRegisters

circuit = QuantumCircuit(2)  # circuit with two qubits
circuit.h(0)  # apply H gate to qubit - puts qubit in superposition

simulator = Aer.get_backend('statevector_simulator')  # create simulator returning state vector of qubit
result = execute(circuit, backend = simulator, shots = 1).result()  # only single shot needed
statevector = result.get_statevector()
print(statevector)  
# j = imaginary unit (i)  so [1, 0] = |0>
array_to_latex(statevector)  # prints state vector => |0> or |1>
plot_bloch_multivector(statevector)  # print bloch sphere
```

```python
# 2 qubits - qubit 0 in superposition - measurement on all - QASM simulator
# For QASM, get_counts and plot_histogram
# Qubit 0 either |0> or |1>, Qubit 1 always |0> so |00> or |01>
# notice no QuantumRegisters

circuit = QuantumCircuit(2)  # circuit with two qubits
circuit.h(0)  # apply H gate to qubit - puts qubit in superposition
circuit.measure_all()

simulator = Aer.get_backend('qasm_simulator')  # create simulator returning state vector of qubit
result = execute(circuit, backend = simulator).result()  # no shots specified (1024 by default)
counts = result.get_counts()
print(counts)  # 
# j = imaginary unit (i)  so [1, 0] = |0>
plot_histogram(counts)  # print bloch sphere
```
