# Implement two bit adder with carry-out (carry over to next place)
### cbit 0 for sum and cbit 1 for carry-out

```python
qr = QuantumRegister(4)
cr = ClassicalRegister(2)

circuit = QuantumCircuit(qr, cr)  # circuit with four qubits
#circuit.x(0)
#circuit.x(1)
circuit.barrier()
circuit.ccx(0, 1, 3)
circuit.cnot(0, 1)
circuit.swap(1, 2)
circuit.barrier()
circuit.measure([2, 3], [0, 1])
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')  # create simulator
result = execute(circuit, backend = simulator, shots = 1).result()  # execute simulator 1024 times
plot_histogram(result.get_counts())  #plot counts of all executions
```
