```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
#initialize 3-qubit circuit with named registers
q0 = QuantumRegister(1, name="q0")  # source (sender) qubit
q1 = QuantumRegister(1, name="q1")  # half of entangled Bell state qubit pari sent to sender
q2 = QuantumRegister(1, name="q2")  # half of entangled Bell state qubit pair retained by reciever (receiver)

crz = ClassicalRegister(1, name="crz") # sender's measurement of source qubit
crx = ClassicalRegister(1, name="crx") # sender's measurement of their half of entangled pair

circuit = QuantumCircuit(q0, q1, q2, crz, crx)

# initialize source qubit (quantum state to teleport)
# circuit.rx(pi/3, q0)
circuit.x(q0)

circuit.barrier()

# create Bell state pair
circuit.h(q1)
circuit.cx(q1, q2)

circuit.barrier()

# sender performs Bell state mercuitasurement
circuit.cx(q0, q1)
circuit.h(q0)
circuit.barrier()
circuit.measure(q0, crz)
circuit.measure(q1, crx)
circuit.barrier()

#receiver transforms their qubit based on sender mesaurement results
circuit.x(q2).c_if(crx, 1) # apply X gate if sender's measurment of entangled qubit is 1
circuit.z(q2).c_if(crz, 1) # apply Z gate if source qubit measures to be 1

circuit.draw(output='mpl')

```

```python
simulator = Aer.get_backend('statevector_simulator')  # create simulator returning state vector of qubit
result = execute(circuit, backend = simulator).result()  # only single shot needed
statevector = result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector = }\n") 
```

```python
plot_bloch_multivector(statevector)  # print bloch sphere
```

```python
cr_result = ClassicalRegister(1, name="result")
circuit.barrier()
circuit.add_register(cr_result)
circuit.measure(2, 2)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')  # create simulator
result = execute(circuit, backend = simulator, shots = 102400).result()  # execute simulator 1024 times
plot_histogram(result.get_counts())  #plot counts of all executions
```
